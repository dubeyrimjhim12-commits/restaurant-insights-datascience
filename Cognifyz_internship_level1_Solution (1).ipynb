1. Loading Data and Identifying Dimensions :
We start by loading the dataset using Pandas. We use the .shape attribute to identify the total number of rows and columns, which gives us an idea of the dataset's size.


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/Dataset .csv')

# Identify rows and columns
rows, cols = df.shape
print(f"The dataset has {rows} rows and {cols} columns.")
df.head() # Displays the first 5 rows
```

2. Handling Missing Values
Missing data can bias our analysis. We check for null values in each column. If missing values are found (commonly in the 'Cuisines' column), we handle them by dropping those rows to ensure data quality.


```python
# Check for missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

# Dropping rows with missing values
df.dropna(inplace=True)

# Verification
print("\nMissing values after cleaning:", df.isnull().sum().sum())
```

3. Data Type Verification
It is essential to ensure each column has the correct data type (e.g., numerical values as floats/ints). This step confirms that our data is ready for mathematical operations.

```python
# Display data types
print("Current Column Data Types:")
print(df.dtypes)
```

4. Distribution of Target Variable ("Aggregate rating")
We analyze the "Aggregate rating" column to see how ratings are distributed. Using a histogram, we can identify Class Imbalance (e.g., if one particular rating like 0.0 dominates the dataset).


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='teal')
plt.title('Distribution of Aggregate Rating')
plt.show()

# Counting occurrences of each rating to check for imbalance
print("Rating Value Counts:")
print(df['Aggregate rating'].value_counts())
```

**Level 1 - Task 2: Descriptive Analysis**

1. Statistical Summary
We use the .describe() method to find the mean, median (50th percentile), standard deviation, and range for all numerical data. This gives us a quick overview of the data's central tendency.

2. Categorical Variable Exploration
In this step, we identify the most frequent categories in our dataset. Specifically, we look for the top cuisines and cities with the highest number of restaurants.


```python
# Identifying top cuisines and cities
print("Top 10 Cities with the most restaurants:")
print(df['City'].value_counts().head(10))

print("\nTop 10 Cuisines in the dataset:")
print(df['Cuisines'].value_counts().head(10))
```

**Level 1 - Task 3: Geospatial Analysis**

Objective: To visualize restaurant locations on a map and investigate if there's any correlation between geographical location and ratings.


1. Visualizing Locations
We use Latitude and Longitude to plot the restaurants on a scatter plot. The color of the points represents the "Aggregate rating," helping us see if highly-rated restaurants are clustered in certain areas.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing restaurant locations
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Longitude', y='Latitude', data=df, hue='Aggregate rating', palette='viridis', alpha=0.5)
plt.title('Geospatial Distribution of Restaurants')
plt.show()
```

2. Correlation between Location and Rating
We check if Latitude and Longitude have any statistical relationship with the restaurant's rating using a correlation matrix.

```python
# Checking for correlation between coordinates and rating
correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print("Correlation Matrix (Location vs Rating):")
print(correlation_matrix)
```
